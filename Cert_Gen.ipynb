# 5. Creating Certificate Logic Function

```python
from re import template
# in this function, we will create a logic of creating a cert generator
# will use this function 11 times

def certificate_generator(name,coursename,courselevel,date):

  pdf_file_name = "/content/drive/MyDrive/Colab Notebooks/Code Ranger Class/1.2 Python Certificate Generator Project (STARTING TEMPLATE)/certificates" + name + coursename + courselevel + date + ".pdf"

  # canvas = blank screen (A4 landscape)
  # we will be storing the blank canvas inside of this pdf

  # we will be storing the canvas in variable called "c"
  # canvas will require 2 inputs:
    # (i) location of the file --> "pdf_file_name"
    # (ii) size of the canvas

  c = canvas.Canvas(pdf_file_name, pagesize= landscape(A4))
  template = "/content/drive/MyDrive/Colab Notebooks/Code Ranger Class/1.2 Python Certificate Generator Project (STARTING TEMPLATE)/certificate_template.jpg"

   # Now we want to draw our cert template to our canvas >> draw image
   # DrawImage() 5 inputs
   # i)  what image you gonna use?
  # ii) x axis (coordinate 0)
  # iii) y axis (coordinate 0)
  # iv) width A4[1] (using the A4 variable we imported)(where to start drawing) (A4[1] standard A4 height)
  # v) height A4[0] (using the A4 variable we imported)(where to start drawing) (A4[0] standard A4 width)

  c.drawImage(template,0,0,A4[1],A4[0]) # to set the location
  c.save()
```

```python
# cert_generator needs 4 inputs
certificate_generator("Jun","Python","Beginner","20/06/2024")
certificate_generator("Veshal","Web Development","Advanced","21/06/2024")
```
