```python
from re import template
# in this function, we will create a logic of creating a cert generator
# will use this function 11 times

def certificate_generator(name,coursename,courselevel,date):

  pdf_file_name = "/content/drive/MyDrive/Colab Notebooks/Code Ranger Class/1.2 Python Certificate Generator Project (STARTING TEMPLATE)/certificates/" + name + "-" + coursename + "-"+ courselevel + ".pdf"

  c = canvas.Canvas(pdf_file_name, pagesize= landscape(A4))
  template = "/content/drive/MyDrive/Colab Notebooks/Code Ranger Class/1.2 Python Certificate Generator Project (STARTING TEMPLATE)/certificate_template.jpg"

  c.drawImage(template,0,0,A4[1],A4[0]) # to set the location

# to create a variable of the middle A4 page
  center_x = c._pagesize[0]/2
  center_y = c._pagesize[1]/2

# Set our font --> .setfont >> 2 input (font type & font size)
# .drawCenteredString >> where you gonna draw the name and what to input >> x axis on the canvas
# i) x axis on the canvas
#ii) what do you want to draw on the canvas

# Name:
  c.setFont('Lora-Bold',30)
  c.drawCentredString(center_x,center_y-40,name)

# Coursename

  c.setFont('Lora-Bold',20)
  c.drawCentredString(center_x,center_y-105,coursename + "-" + courselevel)

# date

  c.setFont('Lora-Bold',15)
  c.drawCentredString(center_x + 190,center_y-160,date)

#Cert ID
  cert_id = "Cert ID " + str(pd.Timestamp.now().timestamp()).replace(".","")
  c.setFont('Lora-Regular',10)
  c.drawCentredString(center_x + 266,center_y-230,cert_id )

  c.save()
```
